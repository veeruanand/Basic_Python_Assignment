```python
#Q2Describe the role of predefined keywords in Python and provide examples of how they are used in a 
#program
#A2Predefined keywords in Python are reserved words that have special meanings and are used to define the syntax and structure of the language

# Example of control flow keywords
for i in range(5):
    if i % 2 == 0:
        print(f"{i} is even")
    else:
        print(f"{i} is odd")
        
        
        
# Example of function definition and return keyword
def add(a, b):
    return a + b

result = add(3, 5)
print(f"Sum: {result}")

# Example of logical operations
a = True
b = False
print(a and b)  
print(a or b)  
print(not a)   
```

```python
#Q3Compare and contrast mutable and immutable objects in Python with examples
#A3Mutable objects are those that can be changed after they are created. This means we can modify their content without changing their identity.
#Example
#Lists, Dictionaries, Set

my_list = [1, 2, 3]
my_list.append(4) 
print(my_list)

my_dict = {'a': 1, 'b': 2}
my_dict['c'] = 3  
print(my_dict)  

my_set = {1, 2, 3}
my_set.add(4) 
print(my_set) 


#Immutable objects, on the other hand, cannot be changed after they are created.
#Ex: Strings, Tuples, Integers

my_string = "hello"
new_string = my_string.replace('h', 'j')  
print(my_string) 
print(new_string)

my_tuple = (1, 2, 3)
new_tuple = (4,) + my_tuple[1:]  
print(my_tuple)  
print(new_tuple)

my_int = 10
new_int = my_int + 5 
print(my_int)  
print(new_int) 
```

```python
#Q4Discuss the different types of operators in Python and provide examples of how they are used
#A4 Arithmatic, Substraction, Multiplication, Division, Modulus,Floor Division , Not equal, Greator Than etc

#Arithmatic
#Addition

x = 5
y = 3
print(x + y)

#Substraction
print(x - y)

#Multiplication 
print(x * y)

#Division
print(x / y)

#Modulus

print(x % y)

#Exponent/Power
print(x ** y)

#Floor Division 
print(x // y)

####Assignment Operators
#Add and Assign
x = 8
x += 3
print(x)

#Subtract and Assign 
x -= 2
print(x)

#Comparison Operators
#Equal
print(x == y)

#Not Equal
print(x != y)

#Greater Than
x = 4
y = 6
print(x > y)


#Bitwise Operators
#AND
x = 5
y = 3
print(x & y) 

#OR
print(x | y)

#XOR
print(x ^ y)

#Left Shift
print(5 << 1)

#Right Shift
print(5 >> 1)
```

```python
#Q5Explain the concept of type casting in Python with examples
#A5Type casting in Python refers to the process of converting a variable from one data type to another
#Example:
# Implicit type casting

a = 5  
b = 2.5 
print(type(a))
print(type(b))
c = a + b 
print(c)  
print(type(c))

#Explicit Type Casting
x = 3.14
print(type(x))
y = int(x)
print(type(y)) 

num = 10
s = str(num)
print(s)  
print(type(s)) 
```

```python
#Q6How do conditional statements work in Python? Illustrate with examples
#A6Conditional statements in Python allow to execute specific blocks of code based on whether a condition is true or false.

#if statement

x = 10
if x > 5:
    print("x is greater than 5")

#if-else statement

x = 3
if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")

#if-elif statement
x = 7
if x > 10:
    print("x is greater than 10")
elif x > 5:
    print("x is greater than 5 but less than or equal to 10")
else:
    print("x is 5 or less")
    
#nested if statement

x = 8
if x > 5:
    if x > 7:
        print("x is greater than 7")
    else:
        print("x is greater than 5 but less than or equal to 7")
else:
    print("x is 5 or less")
```

```python
#Q7Describe the different types of loops in Python and their use cases with examples.
#A7 For Loops and While Loops

#### For Loops
####The for loop is used for iterating over a sequence (such as a list, tuple, string, or range).

fruits = ["grape", "kiwi", "dargonfruit"]
for fruit in fruits:
    print(fruit)

for i in range(5):
    print(i)
    
    
###While Loop
###The while loop is used to execute a block of code repeatedly as long as a given condition is true

i = 1
while i <= 5:
    print(i)
    i += 1

    
    
####Nested Loops
###Nested loops are loops within loops. They are used when need to perform repeated actions within repeated actions.

# Example
for i in range(1, 4):
    for j in range(1, 4):
        print(i * j, end=" ")
    print()
    
###Loop Control Statements
### Used to terminate the Loops

###Break

for i in range(10):
    if i == 5:
        break
    print(i)
    
###Continue
### Skip current iteration

for i in range(5):
    if i == 2:
        continue
    print(i)
```
